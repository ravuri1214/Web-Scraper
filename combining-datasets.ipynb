```python
import pandas as pd
```

```python
df_1 = pd.read_csv('/kaggle/input/books-csv/books.csv')
df_2 = pd.read_csv('/kaggle/input/data-csv/data.csv')
```

```python
df_1.head()
```

```python
df_2.head()
```

```python
df = pd.DataFrame()
```

```python
df['Title'] = df_1['title']
df['Type']  = df_2['Type']
df['product_id']  = df_2['UPC']
df['price_b_tax'] = df_2['price_b_tax']

df['price_a_tax']  = df_2['price_a_tax']
df['tax']    = df_2['tax']
df['Rating']  = df_1['Rating']
df['Reviews']   = df_2['Reviews']
df['Stock']   = df_1['Availability']
df['Availability'] = df_2['Availability']

df['book_link'] = df_1['book_link']
df['img_link']  = df_1['img_link']
```

```python
df
```

```python
df.to_csv('Final.csv', index = 'False')
```
