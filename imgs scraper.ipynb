```python
from selenium.webdriver.chrome.service import Service

from bs4 import BeautifulSoup
import pandas as pd
```

```python
service = Service(r'D:\Anaconda.com\Lib\site-packages\chromedriver_binary\chromedriver_32\chromedriver-win32\chromedriver.exe')
driver = webdriver.Chrome(service = service)
```

```python
driver.get('https://stock-pictures.netlify.app/')
```

```python
soup = BeautifulSoup(driver.page_source, 'html.parser')
```

```python
data = []
for sp in soup.find_all('div', class_ = 'container'):
    link = sp.find('img').get('src')

    one_split  = sp.find('span', class_ = 'tag-color').text.split(',')[0][7:]
    two_split = sp.find('span', class_ = 'tag-color').text.split(',')[1:]
    Tags = one_split + " " + ' '.join(two_split)
    for i in  sp.find('div', class_ = 'likes-comments'):
        likes = i.text.split(' ')[0]
        break
    for i in  sp.find('div', class_ = 'likes-comments'):
        comments = i.text.split(' ')[0]
    data.append([link, Tags, likes, comments])
```

```python
df = pd.DataFrame(data, columns = ['link', 'tags', 'likes', 'comments'])
```

```python
df.to_csv('imgdata.csv', index = False)
```
